```python
import numpy as np
```

```python
arr = np.array([1, 2, 3])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
print(arr_2d)
```

```python
print(np.zeros((3, 3)))
print(np.ones((2, 4)))
print(np.eye(3))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))
print(np.random.rand(3, 2))
print(np.random.randn(3, 2))
```

```python
print(np.random.randint(1,10))
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)
print(np.sin(a))
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.dot(A, B))
#print(A@B)


 
```

```python
arr = np.array([1, 2, 3, 4])
print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.std())
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr[0])
print(arr[1:4])
print(arr[::2])
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr[0, 1])
print(arr[:, 1])
print(arr[1, :2])
```

```python
arr = np.array([1, 2, 3, 4, 5])
mask = arr > 3
print(arr[mask])
```

```python
arr = np.arange(9)
print(arr.reshape(3, 3))
```

```python
a = np.array([1, 2])
b = np.array([3, 4])
print(np.concatenate((a, b)))
```

```python
print(np.vstack((a, b)))
print(np.hstack((a, b)))
```
