```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]          # X-axis values
y = [10, 20, 25, 30]      # Y-axis values

plt.plot(x, y)            # Create a line plot
plt.show()                # Display the plot
```

```python
plt.plot(x, y, color='red', linestyle='-.', marker='s', linewidth=2)
plt.title('Customized Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)   # Adds grid lines
plt.show()
```

```python
x = [1, 2, 3]          

y1 = [1, 4, 9]        
y2 = [1, 2, 3]        

plt.plot(x, y1, color='blue', linestyle='-', marker='o', label='Squares')
plt.plot(x, y2, color='red', linestyle='--', marker='s', label='Linear')
plt.title('Squares vs Linear')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C']   
values = [5, 7, 3]             

plt.bar(categories, values)    
plt.show()                     
```

```python
data = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5)
plt.show()
```

```python
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title('Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 2, 3], color='blue')
plt.title('Increasing Line')

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [3, 2, 1], color='red')
plt.title('Decreasing Line')

plt.suptitle('Subplot Example')   # Overall title
plt.show()
```
